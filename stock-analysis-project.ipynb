Convert date to datetime

```python
dataset['date'] = pd.to_datetime(dataset['date'])
```

 Drop missing values

```python
dataset = dataset.dropna()
```

Sort by date

```python
dataset = dataset.sort_values(by='date')
```

```python
dataset.dtypes
```

Check unique companies

```python
print(dataset['Name'].unique())
```

```python
print((dataset['Name']).nunique())
```

```python
dataset['Name'].value_counts()
```

Select top 3 companies

```python
top_companies = ['AAPL', 'MSFT', 'GOOGL']
```

Filter dataset for only these companies

```python
filtered_data = dataset[dataset['Name'].isin(top_companies)]
```

# Preview

```python
print(filtered_data.head())
```

Create a new column 'daily_return

```python
filtered_data['daily_return'] = (filtered_data['close'] - filtered_data['open']) / filtered_data['open']
```

Check if it worked

```python
print(filtered_data[['date', 'Name', 'open', 'close', 'daily_return']].head(10))
```

```python
print(filtered_data[['date', 'Name', 'open', 'close', 'daily_return']].tail(10))
```

```python
print(filtered_data[['date', 'Name', 'open', 'close', 'daily_return']].describe(include='all'))
```

 Create a copy to avoid warning

```python
data = filtered_data.copy()
```

Sort values by Name and date

```python
data = data.sort_values(by=['Name', 'date'])
```

Calculate 7-day moving average

```python
data['MA_7'] = data.groupby('Name')['close'].transform(lambda x: x.rolling(window=7).mean())
```

Calculate 30-day moving average

```python
data['MA_30'] = data.groupby('Name')['close'].transform(lambda x: x.rolling(window=30).mean())
```

View the result

```python
print(data[['date', 'Name', 'close', 'MA_7', 'MA_30']].head(15))
```

```python
data[['date', 'Name', 'close', 'MA_7', 'MA_30']].tail(15)
```

Group by company

```python
volatility = data.groupby('Name')['daily_return'].std()
```

Convert the result to a DataFrame

```python
volatility_df = pd.DataFrame(volatility).reset_index()
```

Rename the column for clarity

```python
volatility_df.columns = ['Company', 'Volatility']
```

Sort by highest volatility

```python
volatility_df = volatility_df.sort_values(by='Volatility', ascending=False)
```
